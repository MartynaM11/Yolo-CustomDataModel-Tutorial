```python
import os
import random
import shutil
```

```python
data_path = 'degu img'
```

```python
#utworzenie listy par plików obrazów i dokumentów txt
images_labels = []
for filename in os.listdir(data_path):
    if filename.endswith('.jpg'):
        label_file = filename.replace('.jpg', '.txt')
        images_labels.append((filename, label_file))
```

```python
print("Number of connected pairs:",len(images_labels))
```

```python
random.seed(8)
```

```python
random.shuffle(images_labels)
```

```python
images_labels
```

```python
train_el = 0.75
test_el = 0.03
val_el = 0.22
```

```python
num_train = int(len(images_labels) * train_el)
num_test = int(len(images_labels) * test_el)
num_val = int(len(images_labels) * val_el)
print('num_train',num_train)
print('num_test',num_test)
print('num_val',num_val)
```

```python
# Split
train_images_labels = images_labels[:num_train]
test_images_labels = images_labels[num_train:num_train+num_test]
val_images_labels = images_labels[-num_val:]
```

```python
# Create new directories
images = os.path.join(data_path, 'images')
if not os.path.exists(images):
    os.makedirs(images)
    
labels= os.path.join(data_path, 'labels')
if not os.path.exists(labels):
    os.makedirs(labels)
```

```python
# Create new directories
#images
train = os.path.join(images, 'train')
test = os.path.join(images, 'test')
val = os.path.join(images, 'val')

if not os.path.exists(train):
    os.makedirs(train)

if not os.path.exists(test):
    os.makedirs(test)
    
if not os.path.exists(val):
    os.makedirs(val)

#labels
train = os.path.join(labels, 'train')
test = os.path.join(labels, 'test')
val = os.path.join(labels, 'val')

if not os.path.exists(train):
    os.makedirs(train)

if not os.path.exists(test):
    os.makedirs(test)
    
if not os.path.exists(val):
    os.makedirs(val)
```

```python
for image, label in train_images_labels:
    src_img = os.path.join(data_path, image)
    dst_img = os.path.join('degu img/images/train', image)
    src_lbl = os.path.join(data_path, label)
    dst_lbl = os.path.join('degu img/labels/train', label)
    shutil.move(src_img, dst_img)
    shutil.move(src_lbl, dst_lbl)
```

```python
for image, label in test_images_labels:
    src_img = os.path.join(data_path, image)
    dst_img = os.path.join('degu img/images/test', image)
    src_lbl = os.path.join(data_path, label)
    dst_lbl = os.path.join('degu img/labels/test', label)
    shutil.move(src_img, dst_img)
    shutil.move(src_lbl, dst_lbl)
```

```python
for image, label in val_images_labels:
    src_img = os.path.join(data_path, image)
    dst_img = os.path.join('degu img/images/val', image)
    src_lbl = os.path.join(data_path, label)
    dst_lbl = os.path.join('degu img/labels/val', label)
    shutil.move(src_img, dst_img)
    shutil.move(src_lbl, dst_lbl)
```
